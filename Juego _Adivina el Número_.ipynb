#**ADIVINA EL NÚMERO**

```python
# Elige un número
# Pide al usuario que intente adivinarlo
# Tiene 8 intentos
# El programa puede responder 4 cosas distintas:
# 1. Si el nº que dice es <1 o >100 dirá "No está permitido".
# 2. Si el nº es más alto que el nuestro se lo dirá.
# 3. Si el nº es más bajo que el nuestro se lo dirá.
# 4. Si acierta, ¡premio! Y le dirá cuántos intentos ha gastado.

```

```python
from random import randint
```

```python
mi_num = randint(1,100)

name = input('¡Hola! ¿Cómo te llamas?: \n').capitalize()
print(f'Bien {name}, he pensado un número entre 0 y 100.\nTienes 8 intentos para adivinarlo.\n')

su_num = int(input('Di un número: ' ))

intentos = 0
while intentos < 8:
    if su_num in range(1,101):
        if su_num < mi_num:
            print('Incorrecto. Prueba un número más alto.')
            intentos = intentos + 1
            su_num = int(input('Di un número: ' ))
        elif su_num > mi_num:
            print('Incorrecto. Prueba un número más bajo.')
            intentos = intentos + 1
            su_num = int(input('Di un número: ' ))
        else:
            intentos = intentos + 1
            print(f'¡Enhorabuena {name}! Lo has acertado en {intentos} intentos.')
            break  
    else:
        intentos = intentos + 1
        print('Incorrecto. El número debe estar entre 1 y 100.')
        su_num = int(input('Di un número: ' ))
else:
    print(f'Lo siento {name}, no te quedan más intentos. El número era el',mi_num)


```
